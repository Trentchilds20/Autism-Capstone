```python
import numpy as np
import pandas as pd
```

```python
autism = pd.read_csv('train.csv')
autism.head()
```

Drop the columns that don't have enough variance for them to be useful to the model.

```python
autism.drop(columns=['ID','contry_of_res','used_app_before','age_desc','relation','ethnicity'], inplace=True)
autism.head()
```

Round the age column to be neater.

```python
autism.age = autism.age.round(1)
autism.age.head()
```

Change the gender column to male and then change the values to 0 for female and 1 for male.

```python
autism.rename(columns={'gender':'male','austim':'ASD_relative'}, inplace=True)
autism['male'] = autism['male'].replace({'m': 1, 'f': 0})
autism.male.value_counts()
```

Do the same for the jaundice column

```python
autism['jaundice'] = autism['jaundice'].replace({'no':0,'yes':1})
autism.jaundice.value_counts()
```

And again for ASD relative

```python
autism['ASD_relative'] = autism['ASD_relative'].replace({'no':0,'yes':1})
autism.ASD_relative.value_counts()
```

Round the results column

```python
autism.result = autism.result.round(2)
autism.result.head()
```

```python
autism.head()
```

Scale the result and age columns to be between 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
autism['result'] = scaler.fit_transform(autism[['result']])
autism['age'] = scaler.fit_transform(autism[['age']])
autism.head()
```

```python
autism.to_csv('train_scaled.csv', index=False)
```
